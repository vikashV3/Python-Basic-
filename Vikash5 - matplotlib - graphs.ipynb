```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
x=[5,2,4]
y=[2,6,5]

x2=[1,2,1]
y2=[3,4,2]

plt.plot(x,y,label='BYE',linewidth=0.5)
plt.plot(x2,y2,label='HELLO',linewidth=5)
plt.title('HELLO')
plt.xlabel('THIS IS X AXIS')
plt.ylabel('ṭḥīś īś y āxīś')
plt.legend()

plt.grid(True,color='g')#try color='k'
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
x=[5,2,4]
y=[2,6,5]

x2=[1,2,1]
y2=[3,4,2]

plt.bar(x,y,align='center',label="hi")                       #label is of no use *_* but 
plt.bar(x2,y2,color='g',align='center',label='bye')          #without label no use of legend
plt.title('BAR')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')

x=[5,2,4]
y=[2,6,5]

x2=[1,2,1]
y2=[3,4,2]

plt.scatter(x,y,label="hi",color='k' )                       
plt.scatter(x2,y2,color='g',label='bye')          
plt.title('BAR')
plt.legend()
plt.show()
```
